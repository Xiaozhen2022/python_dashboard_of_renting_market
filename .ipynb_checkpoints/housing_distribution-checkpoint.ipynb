## Convert address data to lattitude & longitude values

```python
import pandas as pd
import requests
import json
import numpy as np
```

```python
#Read houses csv
df = pd.read_csv('houses.csv')
print(df['listingAdd'].notnull().sum(axis = 0))
df = df[df['listingAdd'].notnull()]
df['full_add'] = df['listingAdd']+','+df['listingCity']+',BC,Canada'
df['full_add']
```

## Data Wrangling

```python
#View the number NaN in a column
#Fill sth in field listingTitle
#print(df['listingCity'].isnull().sum(axis = 0))
dd=df[df['listingCity'].isnull()]
conditions=[dd['listingTitle'].str.contains('温西|UBC'),dd['listingTitle'].str.contains('本拿比|Burnaby|Brentwood|Deer Lake|metrotown'),\
            dd['listingTitle'].str.contains('列治文'),dd['listingTitle'].str.contains('素里')]
choices=['Vancourver','Burnaby','Richmond','Surrey']
dd['listingCity']=np.select(conditions,choices)
#print(dd[['id','listingTitle','listingCity']])
df['listingCity']=df['listingCity'].fillna(dd['listingCity'])

#Add room_nums field
dd=df[df['listing_room_type'].notnull()]
dd=df[df['listingTime'].notnull()]
dd['room_nums']=dd['listing_room_type'].str.split('；',expand=True)[0]
df=dd
df
```

## Add some fields

```python

#API Call and get the lattitude & longitude values
#API website: https://developer.mapquest.com/documentation/geocoding-api/
#Add lat, lng fields
for i,row in df.iterrows():
    apiAddress = df.at[i,'full_add']
    parameters = {"key":"qGEMqYy02PiSl6Iv88MdQExwSP4Ybzro", "location":apiAddress}
    response = requests.get("http://www.mapquestapi.com/geocoding/v1/address", params=parameters)
    data = json.loads(response.text)['results'][0]['locations'][0]['latLng']
    df.at[i,'lat'] = data['lat']
    df.at[i,'lng'] = data['lng']
df.to_csv(r"houses_geo.csv",encoding='utf-8')
print('Succeed in storing the houses_geo.csv!')
```

## Python Folium - display renting houses resources

```python
import folium
from folium.plugins import MarkerCluster
#map style: OpenStreetMap, Stamen Terrain, Stamen Toner, Stamen Watercolor, CartoDB positron, CartoDB dark_matter
m = folium.Map(location=[49.23736,-123.00804], tiles='OpenStreetMap', zoom_start=10, width='100%', height='100%')
for i,row in df.iterrows():
    lat=df.at[i,'lat']
    lng=df.at[i,'lng']
    des=df.at[i,'listing_rent_price']
    room_nums=df.at[i,'room_nums']
    if room_nums == '1睡房':
       color='red'
    elif room_nums == '2睡房':
       color='blue'
    elif room_nums == '3睡房':
       color='yellow'
    else:
       color='grey'
    folium.Marker(location=[lat,lng], popup=des, icon=folium.Icon(icon='home',color=color)).add_to(m)
#m.save('index.html')
m
```
